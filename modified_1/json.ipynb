```python
import json
from collections import OrderedDict

# Ready for data
body_part = OrderedDict()
upper = OrderedDict()
lower = OrderedDict()

body_part["upper_body"] = upper

upper["1"] = 10
upper["2"] = 20
upper["3"] = 30
upper["4"] = 40
upper["5"] = 50
upper["6"] = 60
upper["7"] = 70

body_part["lower_body"] = lower

lower["8"] = 80
lower["9"] = 90
lower["10"] = 100
lower["11"] = 110
lower["12"] = 120
lower["13"] = 130


# Print JSON
print(json.dumps(body_part, ensure_ascii=False, indent="\t"))
```

```python
# Write JSON
with open('points.json', 'w', encoding="utf-8") as make_file:
    json.dump(body_part, make_file, ensure_ascii=False, indent="\t")
```
